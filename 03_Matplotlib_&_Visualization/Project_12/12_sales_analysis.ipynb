### Data and Output Folder

```python
import os
import pandas as pd

py_path = os.getcwd()

data_folder = os.path.join(py_path, "Data")
output_folder = os.path.join(py_path, "Output")

for folder in [data_folder,output_folder]:
    if not os.path.exists(folder):
        os.mkdir(folder)
        print(f"Created {folder}.")
```

```python
df = pd.read_csv("Data/Sample - Superstore.csv", encoding="windows-1252")

df.shape

df.columns

df["Order Date"] = df["Order Date"].astype("datetime64[ns]")

df.isnull().sum()

category_list = ["Segment","City","Region","Category","Sub-Category"]

df[category_list].nunique()

# print(df["Category"].unique())
# print(df["Sub-Category"].unique())
# print(df["Region"].unique())
# print(df["City"].unique())
# print(df["Segment"].unique())


```

```python

cat_sale = df.groupby("Category")["Sales"].sum().sort_values(ascending=False)

print(cat_sale)

Reg_sale = df.groupby("Region")["Sales"].sum().sort_values(ascending=False)
print(Reg_sale)

seg_sale = df.groupby("Segment")["Sales"].sum().sort_values(ascending=False)
print(seg_sale)

sub_cat_sale = df.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False).head(10)
print(sub_cat_sale)
```
